## 월 - 구간 별로 사고 횟수

### id columns 생성 (key)

```python
ttt['월'] = ttt['월'].astype(int)
ttt2['월'] = ttt2['월'].astype(int)
sss['월'] = sss['월'].astype(int)
```

```python
ttt['id'] = ttt['월'].astype(str) + ttt['구간명']
ttt2['id'] = ttt2['월'].astype(str) + ttt2['구간명']
sss['id'] = sss['월'].astype(str) + sss['구간명']
```

```python
ttt.head(2)
```

```python
ttt2 = ttt2[['id','평균속도(km/h)']]
ttt2.rename(columns = {'평균속도(km/h)':'전일평균속도(km/h)'},inplace=True)
ttt2.head(2)
```

```python
sss = sss[['id','사고횟수']] # 1887 * 4
```

```python
ttt.shape
```

```python
ttt2.shape
```

```python
new = pd.merge(left = ttt , right = ttt2, how = "inner", on = "id")
```

```python
new = new[['월','구간명','평균속도(km/h)','전일평균속도(km/h)','id']]
new # 5162 * 5
```

사고 table이랑은 left outer join 해줌

```python
newthing = pd.merge(left = new, right = sss, how = 'left', on = 'id')
```

```python
newthing
```

```python
newthing['사고횟수'].isnull().sum() 
```

```python
newthing['사고횟수'].fillna(0,inplace=True)
newthing['사고횟수'].value_counts()
```

```python
# 끝 ~
```

```python
newthing.rename(columns = {'평균속도(km/h)':'주간평균속도(km/h)'},inplace=True)
#newthing = newthing[['월','구간명','주간평균속도(km/h)','전일평균속도(km/h)','사고횟수']]
newthing.tail(3)
```

## 좌표추가

```python
code = df[['월','구간','위도','경도']]
```

```python
code['월'] = code['월'].astype(int)
code['id'] = code['월'].astype(str) + code['구간']
```

```python
code = code[['id','위도','경도']]
code.rename(columns = {'위도':'사고지점위도','경도':'사고지점경도'},inplace=True)
code
```

```python
real_newthing = pd.merge(left = newthing, right = code, how = 'left', on = 'id')
```

```python
real_newthing['사고지점위도'].fillna(0,inplace=True)
real_newthing['사고지점경도'].fillna(0,inplace=True)
```

```python
real_newthing
```

```python
real_newthing = real_newthing.drop(['id'],axis=1)
real_newthing
```

```python
real_newthing.to_csv('real_newthing.csv',index=False)
```
