```python
import json

def parse_re(input_re: str):
    try:
        return_dict = json.loads(input_re)
    except json.decoder.JSONDecodeError:
        return_dict = dict()
    finally:
        return return_dict
```

```python
import pandas as pd
from pandas.io.json import json_normalize
import requests
import pprint
import json
```

```python
url = 'http://data.ex.co.kr/openapi/restinfo/restOilList?key=2699999413&type=json'
response = requests.get(url) 
contents = response.text 
json_ob = parse_re(contents)
body = json_ob['list']
dataframe = pd.DataFrame(body) # 전엔 빈 df 만들고 append 해줬었는데
                               # 그러지 않아도 쉽게 그냥.. 리스트를 그대로 데이터프레임 만들어주면 됨
                               # 🤗
```
