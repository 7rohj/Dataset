```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_excel("VDS_구간교통량_1일_1개월_202101.xlsx") 
df2 = pd.read_excel("VDS_구간교통량_1일_1개월_202102.xlsx") 
df3 = pd.read_excel("VDS_구간교통량_1일_1개월_202103.xlsx") 
df4 = pd.read_excel("VDS_구간교통량_1일_1개월_202104.xlsx") 
df5 = pd.read_excel("VDS_구간교통량_1일_1개월_202105.xlsx") 
df6 = pd.read_excel("VDS_구간교통량_1일_1개월_202106.xlsx") 
df7 = pd.read_excel("VDS_구간교통량_1일_1개월_202107.xlsx") 
df8 = pd.read_excel("VDS_구간교통량_1일_1개월_202108.xlsx") 
df9 = pd.read_excel("VDS_구간교통량_1일_1개월_202109.xlsx") 
df10 = pd.read_excel("VDS_구간교통량_1일_1개월_202110.xlsx")
df11 = pd.read_excel("VDS_구간교통량_1일_1개월_202111.xlsx") 
df12 = pd.read_excel("VDS_구간교통량_1일_1개월_202112.xlsx") 
```

## concat 함수로 테이블 합쳐줌

```python
df = pd.concat([df1,df2,df3,df4,df5,df6,df7,df8,df9,df10,df11,df12])
```

```python
df.shape
```

## 전처리 (빈칸인 열 삭제)

교통량 결측값 -1로 되어 있음 삭제시켜줌

```python
mask = df['교통량'].isin([-1])
```

```python
df100 = df[~mask]
df100 = df100.reset_index(drop=True)
df100
```

## 집계일자 월_일로 문자열인덱싱

```python
df100.dtypes
```

```python
df100 = df100.astype({'집계일자':'object'})
```

```python
df100['월/일'] = df100['집계일자'].astype(str).str[4:] # 그냥 .str 붙여주면 에러 남.
df100['월'] = df100['월/일'].astype(str).str[:2]
df100
```

## 콘존ID 별 월별 평균 교통량 구해줌

```python
mont = df100.groupby(["월",'콘존ID']).agg({'교통량':"mean"})
mont
```

```python
mont.to_excel('mont.xlsx')
ttt = pd.read_excel("mont.xlsx")
ttt = ttt.fillna(method='ffill')
ttt
```
