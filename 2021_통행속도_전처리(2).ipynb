```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("VDS_구간통행속도_1일_1개월_202101.csv", encoding='cp949') 
df2 = pd.read_csv("VDS_구간통행속도_1일_1개월_202102.csv", encoding='cp949') 
df3 = pd.read_csv("VDS_구간통행속도_1일_1개월_202103.csv", encoding='cp949') 
df4 = pd.read_csv("VDS_구간통행속도_1일_1개월_202104.csv", encoding='cp949')
df5 = pd.read_csv("VDS_구간통행속도_1일_1개월_202105.csv", encoding='cp949')
df6 = pd.read_csv("VDS_구간통행속도_1일_1개월_202106.csv", encoding='cp949')
df7 = pd.read_csv("VDS_구간통행속도_1일_1개월_202107.csv", encoding='cp949')
df8 = pd.read_csv("VDS_구간통행속도_1일_1개월_202108.csv", encoding='cp949') 
df9 = pd.read_csv("VDS_구간통행속도_1일_1개월_202109.csv", encoding='cp949')
df10 = pd.read_csv("VDS_구간통행속도_1일_1개월_202110.csv", encoding='cp949')
df11 = pd.read_csv("VDS_구간통행속도_1일_1개월_202111.csv", encoding='cp949') 
df12 = pd.read_csv("VDS_구간통행속도_1일_1개월_202112.csv", encoding='cp949')
```

## concat 함수로 테이블 합쳐줌

```python
df = pd.concat([df1,df2,df3,df4,df5,df6,df7,df8,df9,df10,df11,df12])
```

```python
df.shape
```

## 전처리 (빈칸인 열 삭제)

평균속도 결측값 0으로로 되어 있음 삭제시켜줌

```python
mask = df['평균속도'].isin([0])
```

```python
df100 = df[~mask]
df100 = df100.reset_index(drop=True)
df100
```

```python
df100 = df100.drop(columns=['Unnamed: 4'])
```

```python
df100.tail(3)
```

## 집계일자 월_일로 문자열인덱싱

```python
df100.dtypes
```

```python
df100 = df100.astype({'집계일자':'object'})
```

```python
df100['월/일'] = df100['집계일자'].astype(str).str[4:] # 그냥 .str 붙여주면 에러 남.
df100['월'] = df100['월/일'].astype(str).str[:2]
df100
```

## 콘존ID 별 월별 평균 교통량 구해줌

```python
mont = df100.groupby(["월",'콘존ID']).agg({'평균속도':"mean"})
mont
```

```python
mont.to_excel('mont.xlsx')
ttt = pd.read_excel("mont.xlsx")
ttt = ttt.fillna(method='ffill') #
ttt
```
